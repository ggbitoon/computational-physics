```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def random_walk(n):
    
    x = [0]
    y = [0]
    delta_x = []
    delta_y = []
    
    for i in range(n):
        delta_x.append((random.random() - 0.5) * 2)
        delta_y.append((random.random() - 0.5) * 2)
        
        if delta_x[i] == 0 and delta_y[i] == 0:
            
            delta_x.append(0)
            delta_y.append(0)
        
        else:
            
            delta_x.append((delta_x[i])/np.sqrt((delta_x[i])**2+(delta_y[i])**2))
            delta_y.append((delta_y[i])/np.sqrt((delta_x[i])**2+(delta_y[i])**2))
            
        x.append(x[i] + delta_x[i])
        y.append(y[i] + delta_y[i])
        
        data = zip(x, y)
        delta = zip(delta_x, delta_y)
        
    return data, delta
```

```python
def walk_plot(n):
        
    x1, y1 = zip(*random_walk(n)[0])
    
    x2, y2 = zip(*random_walk(n)[0])
    
    x3, y3 = zip(*random_walk(n)[0])
    
    x4, y4 = zip(*random_walk(n)[0])

    plt.plot(x1, y1, color = 'red')
    plt.plot(x2, y2, color = 'green')
    plt.plot(x3, y3, color = 'blue')
    plt.plot(x4, y4, color = 'violet')
    plt.show()
```

```python
walk_plot(1000)
```
