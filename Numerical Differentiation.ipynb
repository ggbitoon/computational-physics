```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cos_func(t):
    return np.cos(t)
def exp_func(t):
    return np.exp(t)
def sq_func(t):
    return t*t
```

```python
def forward_diff(func, h):
    return (func(t + h) - func(t)) / h
def central_diff(func, h):
    return (func(t + h / 2) - func(t - h / 2)) / h
def extra_diff(func, h):
    return (8 * func(t + h / 4) - func(t - h / 4) - func(t + h / 2) - func(t - h / 2)) / (3 / h)
```

```python
t = np.linspace(0, 100, 1001)
```

```python
plt.figure(figsize=(15,5))
plt.plot(t, cos_func(t), color='red')
plt.plot(t, forward_diff(cos_func,0.1), color = 'blue')
plt.plot(t, central_diff(cos_func,0.1), color = 'green')
plt.plot(t, np.gradient(cos_func(t),t), color = 'purple')
plt.plot(t, extra_diff(cos_func,0.1))
```
